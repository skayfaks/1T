Создайте функцию , которая будет применена к столбцу рейтинг методом `apply()`.
Функция должна принимать строку с рейтингом и возвращать:
 - High rating, если рейтинг больше или равен 4.5
 - Middle rating, если рейтинг между 3.8 и 4.5
 - Low rating, если рейтинг меньше 3.8 
Результат работы функции запишите в новый столбец с названием `categorical_rating`

````python
def level_rating(x):
    if x >= 4.5:
        return "High rating"
    elif x < 3.8:
        return "Low rating"
    else:
        return "Middle rating"
````

````python
df['categorical_rating'] = df['rating'].apply(level_rating)
df.head(5)
````

Выведите кол-во уникальных значений в столбцах отсортированных по убыванию на экран:

Шаг 2: Внутри функции есть блок try-except. В блоке try выполняется следующее:
- Функция `pd.to_datetime(date)` преобразует значение `date` в объект типа `datetime`.

````python
def extract_day_of_update(date):         # Функция `pd.to_datetime(date)` преобразует значение `date` в объект типа `datetime`.
    try:                                 # Затем мы используем `.day`, чтобы получить только день месяца из этого объекта типа `datetime`.
        return pd.to_datetime(date).day  # Если успешно выполнены все эти шаги, то возвращаем день месяца.
    except:
        return 'miss_date'               # Если что-то происходит не так return 'miss_date'

df['day_of_update'] = df['last_updated'].apply(extract_day_of_update)
display(df.head(5))
````
